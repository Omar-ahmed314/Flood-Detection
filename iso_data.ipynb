```python
from Isodata import *
from skimage import io
```

```python
img = io.imread('test_image/test1.jpeg')
io.imshow(img)
io.show()

isodata = IsodataClustering(num_clusters=2, max_iterations=20, T=1, merge_threshold=10)
model = isodata.fit(img)
print(model.centroids)
io.imshow(model.labels_)
io.show()
```

```python
X = [1, 2, 3, 5, 6]
K = 3
centroids = X[np.random.choice(len(X), K, replace=False).astype(int)]

distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))

print(distances)

# Assign each point to its closest centroid
labels = np.argmin(distances, axis=0)

print(labels)
```
