```python
import numpy as np
from skimage import io
from Isodata import *
```

```python
img = io.imread('dataset/flooded/400.jpg')
io.imshow(img)
io.show()

isodata = IsodataClusteringV2(num_clusters=2, max_iterations=20, T=1, merge_threshold=20, max_variance=5)
mask = isodata.fit(img)

io.imshow(mask)
io.show()
```
